### üçí Selecionando fatias e valores especificos de um DataFrame

<br>
Um DataFrame √© como uma tabela e toda tabela √© composta por linhas e colunas, tendo isso em mente fica f√°cil entender como selecinr as linhas, as colunas e valores especificos.

O que vamos fazer:
- Selecionar linha (s)
- Selecionar coluna (s)
- Selecionando linhas e colunas
- Selecionar valores especificos
- Selecionando dados com base em seu tipo



<br>
Existem 3 maneiras de selecionar partes de um DataFrame, a primeira √© utilizando o pr√≥prio indice do DataFrame, a segunda √© ultilizando o atributo Data.Frame.loc[ ] e a terceira √© ultilizando o atributo DataFrame.iloc[ ]

<br>

**Vou criar um DataFrame para utilizar como exemplo na explica√ß√µes que vir√£o pela frente**
<br>

```python
salarios_df = pd.read_csv('ds_salaries.csv')
```

#### üçç Selecionando linha (s)

```python
# Selecionando linhas usando os indices do DataFrame
salarios_df[0:1]  # Retorna um DataFrame
```

```python
# Selecionando 1 linha usando o atributo DataFrame.loc[]
salarios_df.loc[0] # Retorna uma Series

# Selecionando mais de 1 linha usando o atributo DataFrame.loc[]
salarios_df.loc[0:3] # Retorna um DataFrame
```

```python
# Selecionando 1 linha usando o atributo DataFrame.iloc[]
salarios_df.iloc[0] # Retorna uma Series 

# O iloc usa a contagem padr√£o do python

# Selecionando mais de 1 linha usando o atributo DataFrame.iloc[]
salarios_df.iloc[0:3] # Retorna um DataFrame
```

#### üçç Selecionando coluna (s)

```python
# Selecionando apenas 1 coluna usando os indices do DataFrame
salarios_df['work_year']  # Retorna uma Series

# Selecionando mais de 1 coluna usando os indices do DataFrame
salarios_df[['work_year', 'experience_level', 'job_title', 'salary_in_usd']]  # Retorna um DataFrame
```

```python
# Selecionando apenas 1 coluna usando o atributo DataFrame.loc[]
salarios_df.loc[:,'work_year']  # Retorna uma Series

# Selecionando mais de 1 coluna usando o atributo DataFrame.loc[]
salarios_df.loc[:, ['work_year', 'experience_level', 'job_title', 'salary_in_usd']]  # Retorna um DataFrame
```

```python
# Selecionando apenas 1 coluna usando o atributo DataFrame.iloc[]
salarios_df.iloc[:,2]  # Retorna uma Series

# Selecionando mais de 1 coluna usando o atributo DataFrame.iloc[]
salarios_df.iloc[:, 0:5]  # Retorna um DataFrame

# utilizando o iloc n√£o √© possivel especificar quais colunas voc√™ deseja, o que n√£o √© muito bom.
```

#### üçç Selecionando linhas e colunas

```python
# Selecionando linhas e colunas usando o indice os indices do DataFrame
salarios_df[['work_year', 'job_title']][0:5] # Retorna um DataFrame
```

```python
# Selecionando linhas e colunas usando o atributo DataFrame.loc[]
salarios_df.loc[0:4, ['work_year', 'job_title']]  # Retorna um DataFrame
```

```python
# Selecionando linhas e colunas usando o atributo DataFrame.iloc[]
salarios_df.iloc[0:5, 0:3]  # Retorna um DataFrame

# Com o atributo iloc n√£o √© possivel especificar a coluna
```

#### üçç Selecionando valores especificos

```python
# Selecionando valores especificos usando os indices do DataFrame
salarios_df['work_year'][0]
```

```python
# Selecionando valores especificos usando o atributo DataFrame.loc[]
salarios_df.loc[0,['work_year']]
```

```python
# Selecionando valores especificos usando o atributo DataFrame.iloc[]
salarios_df.iloc[0,1] # significa o primeiro valor da segunda coluna, pois funciona com base na contagem do python.
```

#### üçç Selecionando dados com base em seu tipo

```python
salarios_df.select_dtypes("float") # Retorna somente as colunas que possuirem valores do tipo float

# Nesse df n√£o temos nehum dado desse tipo, por isso ele retorna somente os indices das linhas
```

```python
salarios_df.select_dtypes("object") # Retorna as colunas que possuirem valores do tipo str (object no pandas)
```
